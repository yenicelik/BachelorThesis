## 2D $f : (x_1, x_2) -> y$ with effective dimensionality $d=1$ and $D=2$

```python
%autoreload
Bbf = BBF(1, 2)
inp = np.asarray([1, 1])
print(Bbf.eval_hd(inp))

# Create points to be randomly sampled
samples = 200000
x = (np.random.rand(samples, 2) - 0.4) * 7
y = np.asarray([Bbf.eval_hd(x[i, :]) for i in range(samples)])
print(y.shape)

fig = plt.figure()
ax = fig.gca(projection='3d')
ax.plot_trisurf(x[:,0], x[:,1], y, linewidth=0.01)
plt.show()
```
